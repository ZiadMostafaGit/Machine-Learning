```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import MinMaxScaler,PolynomialFeatures
import seaborn as sns 
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import LabelEncoder
```

```python
new_df=pd.read_csv("cleaned.csv")
```

```python
Y=new_df["SalePrice"]
```

```python
new_df=new_df.drop(columns="SalePrice")
```

```python
correlation_matrix = new_df.corr()
```

```python
#ast column is the target variable
target_variable_name = new_df.columns[-1]

# Get the correlation of each feature with the target variable
correlation_with_target = correlation_matrix[target_variable_name].drop(target_variable_name)

# Sort correlations by absolute value
sorted_correlation_with_target = correlation_with_target.abs().sort_values(ascending=False)
```

```python

# Select the top 10 most correlated features
top_features = sorted_correlation_with_target.head(40).index
```

```python
new_df=new_df[top_features]
```

```python
new_df.info()
```

```python
Y=pd.DataFrame(Y)
Y.shape
```

```python
new_df,test_df,Y,Y_test=train_test_split(new_df,Y,test_size=0.3,random_state=1)
```

```python
print(new_df.shape)
print(test_df.shape)
print(Y.shape)
print(Y_test.shape)
```

```python
model=LinearRegression()
```

```python
scaler=MinMaxScaler()
```

```python
def minMaxTransform(df):
    processor=MinMaxScaler()
    return processor.fit_transform(df),processor
```

```python
poly=PolynomialFeatures(degree=3)
```

```python
new_df=poly.fit_transform(new_df)
test_df=poly.fit_transform(test_df)
```

```python
new_df_scaled,minMaxScalee=minMaxTransform(new_df)
```

```python
test_df_scaled=minMaxScalee.fit_transform(test_df)
```

```python
Y_scaled,processor=minMaxTransform(Y)

```

```python
model.fit(new_df_scaled,Y_scaled)
```

```python
Y_pre=model.predict(test_df_scaled)
```

```python
Y_pre = processor.inverse_transform(Y_pre)
```

```python
mse=mean_squared_error(Y_test,Y_pre)/2
```

```python
r2=r2_score(Y_test,Y_pre)
```

```python
print("\nMean Squared Error (MSE):", mse)
print("R-squared (R2) Score:", r2)
```
